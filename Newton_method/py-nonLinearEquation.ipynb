```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open ("1e-3.txt") as f:
    f1 = f.readlines()
with open ("1e-6.txt") as f:
    f2 = f.readlines()
with open ("1e-9.txt") as f:
    f3 = f.readlines()
```

```python
def getX(file):
    x = []
    for line in file:
        x.append(float(line[:-1]))
    return x\

def func(x):
    return 4*np.sin(x/2) + np.cos(x) * np.tanh(x) - x + 2
```

```python
x1 = getX(f1)
x2 = getX(f2)
x3 = getX(f3)
root = 4.7927776110652285104
```

```python
x0 = np.linspace(0, 10, 100)
y0 = [func(x) for x in x0]
y1 = [func(x) for x in x1]
y2 = [func(x) for x in x2]
y3 = [func(x) for x in x3]
```

```python
plt.figure(figsize =(12, 12))

plt.plot(x0, y0)
plt.plot(x0, [0]*len(x0), 'k')
plt.plot(x1, [0]*len(x1), 'ro')
plt.plot(x1, y1, 'ro')
plt.plot(root, 0, 'go')
for i in range(len(x1) - 1):
    plt.plot(x1[i:i+2], [func(x1[i]), 0], 'k--')
    plt.plot([x1[i]]*2, [func(x1[i]), 0], 'k--')
    
yl = [ 5 - i*20/100 for i in range(100) ]
plt.plot([2]*100, yl, 'g', linewidth = 5)
plt.plot([6.1]*100, yl, 'g', linewidth = 5)

plt.axis([1.5, 6.5, -3, 3.5])
plt.grid()
plt.savefig('foo.jpg')
plt.show()
```
